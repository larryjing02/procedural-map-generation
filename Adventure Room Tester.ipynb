```python
from collections import deque
import random
```

```python
xMax, yMax = 5, 5
minRooms = 15
```

```python
# Generic room class using pointer method to connect between rooms
class Room:
    def __init__(self, name = "Default Room"):
        self.name = name
        self.up = None
        self.down = None
        self.right = None
        self.left = None

    def allowed(self):
        dirs = ""
        if self.up:
            dirs += "W "
        if self.left:
            dirs += "A "
        if self.down:
            dirs += "S "
        if self.right:
            dirs += "D "

        if len(dirs) == 0:
            print("There is nowhere to move in this room!")
        else:
            print("The directions you can move are: " + dirs)

    # Extra helper function with polymorphism - used only during room creation
    def allowed(self, x, y):
        global xMax, yMax
        dirs = []
        if not self.up and y > 0:
            dirs.append(((x, y-1), "W"))
        if not self.left and x > 0:
            dirs.append(((x-1, y), "A"))
        if not self.down and y < yMax:
            dirs.append(((x, y+1), "S"))
        if not self.right and x < xMax:
            dirs.append(((x+1, y), "D"))
        return dirs

    def connect(self, dir, otherRoom):
        # Room A
        # curRoom.connect("W", RoomB)
        if dir == "W":
            self.up = otherRoom
            otherRoom.down = self
        elif dir == "S":
            self.down = otherRoom
            otherRoom.up = self
        elif dir == "A":
            self.left = otherRoom
            otherRoom.right = self
        else:
            self.right = otherRoom
            otherRoom.left = self

    def hi(self):
        print(f"You are currently in Room {self.name}")
```

```python
def drawMap(visited):
    global xMax, yMax
    print("////////////////////////")
    for cx in range(xMax + 1):
        line = ""
        
        for cy in range(yMax + 1):
            if (cx, cy) in visited:
                line += "[ ]"
            else:
                line += "   "
            line += " "
        print(line)
    print("////////////////////////")
```

```python
def generateRandomMap(minRooms):
    random.seed()

    visited = {}
    rooms = deque()


    firstRoom = Room("Starting Room")
    visited[(0,0)] = firstRoom
    rooms.append(((0,0),firstRoom))

    numRooms = 1

    while rooms:
        data = rooms.popleft()
        curRoom = data[1]
        x,y = data[0][0], data[0][1]

        dirs = curRoom.allowed(x,y)
        random.shuffle(dirs)

        numPaths = random.randint(0, len(dirs)-1)
        # Ensure we get at least minRooms rooms
        if numRooms < minRooms:
            numPaths = max(numPaths, 1)

        for d in dirs[0:numPaths]:
            if d[0] in visited:
                curRoom.connect(d[1], visited[d[0]])
            else:
                newRoom = Room(f"Generated: {num}")
                curRoom.connect(d[1], newRoom)
                visited[d[0]] = newRoom
                rooms.append((d[0],newRoom))
                numRooms += 1
            
    drawMap(visited)
```
